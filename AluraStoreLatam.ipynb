```python
# TIENDA 1

# Creación de una lista para las categorias y otra lista de precio conservando el orden de las filas
categoria_lista = [categoria for categoria in tienda1['Categoría del Producto']]
print(f'lista de categorias: {categoria_lista}')
print(f'tamaño categoria: {len(categoria_lista)}')

precios= [precio for precio in tienda1['Precio']]
print(f'lista de precios: {precios}')
print(f'tamaño precios: {len(precios)}')

# Creación de lista de tuplas, porque permite repetir categorias, cosa que los diccionarios no permite repetir claves (Categorias)
categoria_precio = [(cat, valor) for cat, valor in zip(categoria_lista, precios)]
print(f'Lista de tuplas (categoria, precio) es: {categoria_precio}')

# Creación de diccionario vacio donde la clave sea la categoria y su valor la suma del precio de los produtos
categoria_valor1 = {}

# Creacion de una lista con las categorias únicas
cat_unica = list(set(categoria_lista))
print(f'Lista de categorias unica{cat_unica}')

# Agrupación por categoría
for cat_unica in cat_unica:
  for cat, valor in categoria_precio:
    if cat_unica == cat:
      if cat_unica in categoria_valor1:
       categoria_valor1[cat_unica] += int(valor)
      else:
       categoria_valor1[cat_unica] = int(valor)

print(f'La suma de los precios agrupados por categoria es: {categoria_valor1}')
```

# 4. Productos más y menos vendidos

```python
# Creación de un diccionario "producto:unidades_vendidas" por cada tienda
productos_vendidos_tienda1 = {}

for producto in tienda1['Producto']:
    if producto in productos_vendidos_tienda1:
      productos_vendidos_tienda1[producto] += 1
    else:
        productos_vendidos_tienda1[producto] = 1

print(f'productos vendidos en la tienda 1: {productos_vendidos_tienda1}')

productos_vendidos_tienda2 = {}

for producto in tienda2['Producto']:
    if producto in productos_vendidos_tienda2:
      productos_vendidos_tienda2[producto] += 1
    else:
        productos_vendidos_tienda2[producto] = 1

print(f'productos vendidos en la tienda 2: {productos_vendidos_tienda2}')

productos_vendidos_tienda3 = {}

for producto in tienda3['Producto']:
    if producto in productos_vendidos_tienda3:
      productos_vendidos_tienda3[producto] += 1
    else:
        productos_vendidos_tienda3[producto] = 1

print(f'productos vendidos en la tienda 3: {productos_vendidos_tienda3}')

productos_vendidos_tienda4 = {}

for producto in tienda4['Producto']:
    if producto in productos_vendidos_tienda4:
      productos_vendidos_tienda4[producto] += 1
    else:
        productos_vendidos_tienda4[producto] = 1

print(f'productos vendidos en la tienda 4: {productos_vendidos_tienda4}')

```

```python
# Hallar el mayor valor de ventas del diccionario
valor_max_tienda1 = max(productos_vendidos_tienda1.values())
valor_max_tienda2 = max(productos_vendidos_tienda2.values())
valor_max_tienda3 = max(productos_vendidos_tienda3.values())
valor_max_tienda4 = max(productos_vendidos_tienda4.values())

# Filtrar los productos cuya cantidad de ventas concidan con el valor calculado anteriormente
max_venta_tienda1 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda1.items() if num_ventas == valor_max_tienda1 }
max_venta_tienda2 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda2.items() if num_ventas == valor_max_tienda2 }
max_venta_tienda3 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda3.items() if num_ventas == valor_max_tienda3 }
max_venta_tienda4 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda4.items() if num_ventas == valor_max_tienda4 }

print(f'los productos mas vendidos en la tienda 1 son: {max_venta_tienda1}')
print(f'los productos mas vendidos en la tienda 2 son: {max_venta_tienda2}')
print(f'los productos mas vendidos en la tienda 3 son: {max_venta_tienda3}')
print(f'los productos mas vendidos en la tienda 4 son: {max_venta_tienda4}')

# Creacion de diccionario generaal donde cada clave sea el nombre de la tienda y el valor sea el diccionario de sus productos más vendidos.

mas_vendidos = {'tienda 1' : max_venta_tienda1, 'tienda 2' : max_venta_tienda2, 'tienda 3' : max_venta_tienda3, 'tienda 4' : max_venta_tienda4}

# Crear listas para los datos
tiendas = []
productos = []
ventas = []

for tienda, productos_dict in mas_vendidos.items():
    for producto, venta in productos_dict.items():
        tiendas.append(tienda)
        productos.append(producto)
        ventas.append(venta)

# Crear gráfico
plt.figure(figsize=(10, 6))
bars = plt.barh(productos, ventas, color='lightsalmon')

# Añadir texto a las barras
for bar, venta, tienda in zip(bars, ventas, tiendas):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{tienda}', va='center', fontsize=9)

plt.xlabel('Unidades vendidas')
plt.title('Productos más vendidos por tienda')
plt.tight_layout()
plt.show()


```

```python
# Hallar el menor valor de ventas del diccionario
valor_min_tienda1 = min(productos_vendidos_tienda1.values())
valor_min_tienda2 = min(productos_vendidos_tienda2.values())
valor_min_tienda3 = min(productos_vendidos_tienda3.values())
valor_min_tienda4 = min(productos_vendidos_tienda4.values())

# Filtrar los productos cuya cantidad de ventas concidan con el valor calculado anteriormente
min_venta_tienda1 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda1.items() if num_ventas == valor_min_tienda1 }
min_venta_tienda2 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda2.items() if num_ventas == valor_min_tienda2 }
min_venta_tienda3 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda3.items() if num_ventas == valor_min_tienda3 }
min_venta_tienda4 = {producto: num_ventas for producto, num_ventas in productos_vendidos_tienda4.items() if num_ventas == valor_min_tienda4 }

print(f'los productos menos vendidos en la tienda 1 son: {min_venta_tienda1}')
print(f'los productos menos vendidos en la tienda 2 son: {min_venta_tienda2}')
print(f'los productos menos vendidos en la tienda 3 son: {min_venta_tienda3}')
print(f'los productos menos vendidos en la tienda 4 son: {min_venta_tienda4}')


# Creación de diccionario general "menos_vendidos" donde cada clave sea el nombre de la tienda y el valor sea el diccionario de sus productos más vendidos.

menos_vendidos = {'tienda 1' : min_venta_tienda1, 'tienda 2' : min_venta_tienda2, 'tienda 3' : min_venta_tienda3, 'tienda 4' : min_venta_tienda4}

# Crear listas para los datos
tiendas = []
productos = []
ventas = []

for tienda, productos_dict in menos_vendidos.items():
    for producto, venta in productos_dict.items():
        tiendas.append(tienda)
        productos.append(producto)
        ventas.append(venta)

# Crear gráfico
plt.figure(figsize=(10, 6))
bars = plt.barh(productos, ventas, color='lightcyan')

# Añadir texto a las barras
for bar, venta, tienda in zip(bars, ventas, tiendas):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{tienda}', va='center', fontsize=9)

plt.xlabel('Unidades vendidas')
plt.title('Productos menos vendidos por tienda')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# TIENDA 1
# Creación de una lista con los costos de envio por producto de cada tienda
tiendas = [tienda1, tienda2, tienda3, tienda4]

for i, tienda in enumerate(tiendas, start=1):
  lista_envios = tienda['Costo de envío']
  if i == 1:
    promedio_envio_tienda1 = round(sum(lista_envios)/len(lista_envios))
    print(f'El promedio de evio de la tienda {i} es: {promedio_envio_tienda1}')
  elif i == 2:
    promedio_envio_tienda2 = round(sum(lista_envios)/len(lista_envios))
    print(f'El promedio de evio de la tienda {i} es: {promedio_envio_tienda2}')
  elif i == 3:
    promedio_envio_tienda3 = round(sum(lista_envios)/len(lista_envios))
    print(f'El promedio de evio de la tienda {i} es: {promedio_envio_tienda3}')
  else:
    promedio_envio_tienda4 = round(sum(lista_envios)/len(lista_envios))
    print(f'El promedio de evio de la tienda {i} es: {promedio_envio_tienda4}')

Promedio_envio = [promedio_envio_tienda1, promedio_envio_tienda2, promedio_envio_tienda3, promedio_envio_tienda4]
suma_promedio = 0
for promedio in Promedio_envio:
  suma_promedio += promedio
print(f'suma de promedios es: {suma_promedio}')
for i, promedio in enumerate(Promedio_envio, start=1):
  if i == 1:
    porcentaje_envio_tienda1 = ((promedio*100)/suma_promedio)
    print(f'procentaje de envio de la tinda 1 es: {round(porcentaje_envio_tienda1, 1)}')
  elif i ==2:
    porcentaje_envio_tienda2 = ((promedio*100)/suma_promedio)
    print(f'procentaje de envio de la tinda 2 es: {round(porcentaje_envio_tienda2, 1)}')
  elif i ==3:
    porcentaje_envio_tienda3 = ((promedio*100)/suma_promedio)
    print(f'procentaje de envio de la tinda 3 es: {round(porcentaje_envio_tienda3, 1)}')
  else:
    porcentaje_envio_tienda4 = ((promedio*100)/suma_promedio)
    print(f'procentaje de envio de la tinda 4 es: {round(porcentaje_envio_tienda4, 1)}')




```
